```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
dados = pd.read_csv('train.csv')
```

```python
#Identificando valores nulos
dados.isnull().sum().sum()
```

```python
dados["TARGET"].value_counts()
```

```python
#escolhendo as colunas que tem correlação absoluta maior que 0.01
dadoscorr = dados.corr()
```

```python
#escolhendo as colunas que tem correlação absoluta maior que 0.01
dadoslimpos=dados[dadoscorr[(abs(dadoscorr)>0.01)['TARGET']].index]
```

```python
#separando entre target e features
y=dadoslimpos['TARGET']
X=dadoslimpos.drop('TARGET',1)
```

```python
#normalizando os resultados
scaler = MinMaxScaler()
scaler.fit(X)
Xnorm=scaler.transform(X)
```

```python
#Slicing do dataset
X_train, X_test, y_train, y_test = train_test_split(
Xnorm, y, test_size=0.3, random_state=42)
```

```python
#aplicando o Modelo
clf = LogisticRegression(random_state=0,max_iter=1000).fit(X_train, y_train)
```

```python
#Avaliando o Modelo
clf.score(X_test, y_test)
```
